```python
import pandas as pd

first_stage = pd.read_csv("files/aero_first_stage.csv")
second_stage= pd.read_csv("files/aero_second_stage.csv")

first_stage_off = first_stage[["Mach", "CD Power-Off"]].loc[0:2000]
first_stage_on  = first_stage[["Mach", "CD Power-On"]].loc[0:2000]

second_stage_off = second_stage[["Mach", "CD Power-Off"]].loc[0:2000]
second_stage_on  = second_stage[["Mach", "CD Power-On"]].loc[0:2000]

# Export the data frames to the files folder
first_stage_off.to_csv("files/first_stage_off.csv", index=False)
first_stage_on.to_csv("files/first_stage_on.csv", index=False)
second_stage_off.to_csv("files/second_stage_off.csv", index=False)
second_stage_on.to_csv("files/second_stage_on.csv", index=False)
```
